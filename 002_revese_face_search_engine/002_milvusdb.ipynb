````python
from face_recognition import face_encodings, load_image_file
from pathlib import Path
import matplotlib.pyplot as plt
from milvus import Milvus, MetricType
````

````python
image_path = "photos/hinton_0.jpg"
image_array = load_image_file(image_path)
face_embedding = face_encodings(image_array)[0]
print("face_embedding :: ", face_embedding)
print("shape of face_embedding :: ", face_embedding.shape)
print("data type of face_embedding :: ", face_embedding.dtype)
plt.imshow(image_array)
plt.axis('off')
plt.show()
````

--------------------------------------------------------------------------------------------------------------

````python
host = 'localhost'
port = '19530'
client = Milvus(host, port)
````

````python
collection_name = 'face_tutorial'
collection_param = {
     "collection_name": collection_name,
     "dimension": 128,
     "index_file_size": 2048,
     "metric_type": MetricType.L2
 }
client.create_collection(collection_param)
````

````python
status, list_of_collections = client.list_collections()
print("status is :: ", status)
print("list of collections is :: ", list_of_collections) 
````

````python
id_to_file_names = {i: str(j) for i, j in enumerate(Path("photos/").glob("*"))}
id_to_file_names
````

````python
%%time
face_embeddings = []
for _id, path in id_to_file_names.items():
    embedding = face_encodings(load_image_file(path))
    face_embeddings.append(list(embedding[0]))
        
status, ids = client.insert(collection_name, face_embeddings, list(id_to_file_names.keys()))
print("status:: ", status)
print("ids:: ", ids)
````

````python
for unseen_photo_path in Path("unseen_photos/").glob("*"):
    print(80*"-")
    print(">>>> search item ::", unseen_photo_path)
    image_array = load_image_file(unseen_photo_path)
    plt.figure()
    plt.imshow(image_array)
    plt.axis('off')
    plt.show()
    
    face_embedding = face_encodings(image_array)[0]
    status , search_result = client.search(collection_name, 5, [list(map(float, face_embedding))])
    
    plt.figure(figsize=(20,10))
    columns = 5
    print(">>>> search results :: ")
    for i, result in enumerate(search_result[0]):
        ax = plt.subplot(1, columns, i + 1)
        filename = id_to_file_names[result.id]
        ax.set_title(f"{filename}\ndistance={result.distance}")
        plt.imshow(load_image_file(filename))
        plt.axis('off')
    plt.show()


````
